```r
par(bg='white')
func <- function(m, b) {
    # Generate random data
    x <- seq(1, 10, by = 0.1)
    y <- m * x + b + rnorm(length(x), mean = 0, sd = 1)
    # Fit a line to the data
    fit <- lm(y ~ x)
    
    # Plot the data and the line with white background
    plot(x, y, main = "Data and Fitted Line", xlab = "x", ylab = "y")
    abline(fit, col = "red")
}

# Call the function to generate and plot the data with white background
params<-data.frame(m = c(1, 2, 3), b = c(0, 1, 2))
for(row in seq_len(nrow(params))) {
    func(params[row,1], params[row,2])
}



```
